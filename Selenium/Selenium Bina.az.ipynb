# Data Collection

```python
from selenium import webdriver

def post_description(num):
    driver = webdriver.Chrome()
    url = f'https://bina.az/alqi-satqi?page={num}'
    driver.get(url)
    titles = []
    owners = []
    proerties1 = []
    decription = []
    pice_meters = []
    locations = []
    prices = []
    home_links = []
    elements_links = driver.find_elements("css selector", "a.item_link")
    for x in elements_links:
        home_links.append(x.get_attribute('href'))
    # Making links unique because most of them are repeated
    unique_list = list(dict.fromkeys(home_links))
    home_links = unique_list

    for x in home_links:
        try:
            driver.get(x)  # Reuse the existing driver for each post
            
            #Prices
            price = driver.find_element("css selector", "div.product-price__i--bold")
            prices.append(price.text)
            
            #Location
            location = driver.find_element("css selector", "li.product-extras__i")
            locations.append(location.text)
            

            # Owners
            product_owner = driver.find_element("css selector", "div.product-owner__info")
            owner = product_owner.text.split('\n')
            owners.append(owner)

            # Description
            desc_post = driver.find_element("css selector", "div.product-description__content")
            description.append(desc_post.text)

            # Price_meter
            price_m = driver.find_element("css selector", "div.product-price")
            price_meter = price_m.text.split('\n').pop()
            price_meters.append(price_meter)

            # Titles
            title = driver.find_element("css selector", "h1.product-title")
            titles.append(title.text)

            # Properties
            properties = driver.find_elements("css selector", "span.product-properties__i-value")
            property = []
            for p in properties:
                property.append(p.text)
            properties1.append(property)

        except Exception as e:
            print(f"Error occurred for URL: {url}\nError message: {str(e)}")
    
    driver.quit()  # Close the webdriver instance after all posts have been processed
    Converting to Dataframe
    df1 = pd.DataFrame(properties1, columns=['Category','Floor','Area','Rooms','Bill_of_sale','Renovation','Id'])
    df1 = df1.drop('Id',axis=1)
    df1['Description'] = description
    df1['Owners'] = owners
    df1['Titles'] = titles
    df1['Price_meter'] = price_meters
    df1['Links'] = home_links
    df1['Location'] = locations
    df1['Prices'] = prices
    
    
    
    return df1
```
