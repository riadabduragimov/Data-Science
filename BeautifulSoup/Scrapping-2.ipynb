```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
url = 'https://kontakt.az/telefonlar/mobil-telefonlar/'
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, "lxml")
phone_divs = soup.find_all("div",{'class':'name'}) #Information about phones
phone_divs
```

```python
for product in phone_divs: 
    productlink = product.find_all("a",attrs={"class": None})
    for product in productlink:
        print(product.text) #printing all models of phones that are available
        
```

```python
for product in phone_divs: 
    productlink = product.find_all("a",attrs={"class": None})
    for product in productlink:
        print(product.get('href')) #printing all links of phone models
```
