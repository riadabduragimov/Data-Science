```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
RFM = df[['Customer ID','InvoiceDate','Quantity','Price']]
```

```python
RFM
```

```python
NOW = dt.datetime(2010,12,9)
RFM_table=RFM.groupby('Customer ID').agg({'InvoiceDate': lambda x: (NOW - x.max()).days, # Recency
                                                'Quantity': lambda x: len(x.unique()), # Frequency
                                                'Price': lambda x: x.sum()})    # Monetary 
```

```python
RFM_table['InvoiceDate'] = RFM_table['InvoiceDate'].astype(int)
```

```python
RFM_table.rename(columns={'InvoiceDate': 'recency', 
                         'Quantity': 'frequency',
                         'Price': 'monetary_value'}, inplace=True)
```

```python
RFM_table.head()
```

```python
quantiles = RFM_table.quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
```

```python
RFM_Segment = RFM_table.copy()
```

```python
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def R_Class(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
    
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FM_Class(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
```

```python
RFM_Segment['R_Quartile'] = RFM_Segment['recency'].apply(R_Class, args=('recency',quantiles,))
RFM_Segment['F_Quartile'] = RFM_Segment['frequency'].apply(FM_Class, args=('frequency',quantiles,))
RFM_Segment['M_Quartile'] = RFM_Segment['monetary_value'].apply(FM_Class, args=('monetary_value',quantiles,))
RFM_Segment['RFMClass'] = RFM_Segment.R_Quartile.map(str) \
                            + RFM_Segment.F_Quartile.map(str) \
                            + RFM_Segment.M_Quartile.map(str)
```

```python
RFM_Segment
```
